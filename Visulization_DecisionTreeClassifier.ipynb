## **Creating a Decison Tree Classifier and Visualizing it Graphically**
Sir Ronald Fisher's IRIS dataset is being used for this classification algorithm.


---




```
#Importing all the libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
%matplotlib inline
```

```
# importing the IRIS dataset
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names) #creating the IRIS dataframe
df.head()
```

```
y = iris.target #the classification target i.e the dependent features
print(y) 
```

```
#creating the decision tree classifier model
model = tree.DecisionTreeClassifier()
model = model.fit(iris.data , y= iris.target) #iris.data represents the independent features
```

### **Graphically Visualizing the decision tree:**


```
plt.figure(figsize=(15,10))
tree.plot_tree(model, filled=True)
```

**New test data can be fed to this classifer and it would be able to predict the right class accordingly.**
